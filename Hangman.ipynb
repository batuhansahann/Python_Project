```python
import random

# List of possible words for the hangman game
word_list = ['python', 'development', 'hangman', 'challenge', 'computer', 'programming', 'algorithm']

def get_random_word(word_list):
    """
    Function to get a random word from the word list
    """
    return random.choice(word_list)

def display_word(word, guessed_letters):
    """
    Function to display the word with underscores for letters not yet guessed
    """
    return ' '.join([letter if letter in guessed_letters else '_' for letter in word])

def play_hangman():
    """
    Main function to play the hangman game
    """
    print("Welcome to Hangman!")
    
    word = get_random_word(word_list)  # Select a random word
    guessed_letters = set()           # Keep track of guessed letters
    attempts = 6                      # Number of allowed wrong attempts
    correct_guesses = set(word)       # Letters in the word that need to be guessed

    while attempts > 0 and correct_guesses:
        print(f"\nWord to guess: {display_word(word, guessed_letters)}")
        print(f"Attempts remaining: {attempts}")
        guess = input("Guess a letter: ").lower()

        if guess in guessed_letters:
            print("You've already guessed that letter. Try again.")
        elif guess in word:
            print(f"Good guess! '{guess}' is in the word.")
            guessed_letters.add(guess)
            correct_guesses.discard(guess)
        else:
            print(f"Wrong guess! '{guess}' is not in the word.")
            guessed_letters.add(guess)
            attempts -= 1

        if not correct_guesses:
            print(f"\nCongratulations! You guessed the word: {word}")
            break
        elif attempts == 0:
            print(f"\nGame Over! The word was: {word}")

play_hangman()
```
