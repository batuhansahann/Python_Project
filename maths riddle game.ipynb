```python
import random
import operator

def display_welcome():
    print("Welcome to the Math Quiz Game!")
    print("You will be asked random math questions.")
    print("Answer them correctly to reach the target score and win the game.\n")

def generate_question():
    # Define operations and their corresponding symbols
    operations = {
        "+": operator.add,
        "-": operator.sub,
        "*": operator.mul,
        "/": operator.truediv
    }
    
    # Randomly choose two numbers and an operation
    num1 = random.randint(1, 10)
    num2 = random.randint(1, 10)
    operation_symbol = random.choice(list(operations.keys()))
    
    # Ensure no division by zero
    if operation_symbol == "/" and num2 == 0:
        num2 = 1
    
    # Calculate the correct answer using the chosen operation
    answer = operations[operation_symbol](num1, num2)
    
    # Format answer for integer results when appropriate
    if operation_symbol == "/":
        answer = round(answer, 2)  # Round to 2 decimal places if division
    
    question = f"What is {num1} {operation_symbol} {num2}?"
    return question, answer

def play_game():
    display_welcome()
    target_score = 5  # Number of correct answers needed to win
    score = 0

    while score < target_score:
        question, correct_answer = generate_question()
        print(question)
        
        try:
            user_answer = float(input("Your answer: "))
            
            if user_answer == correct_answer:
                score += 1
                print(f"Correct! Your current score is: {score}/{target_score}\n")
            else:
                print(f"Incorrect. The correct answer was: {correct_answer}. Try another question!\n")
        
        except ValueError:
            print("Invalid input. Please enter a numerical answer.\n")
    
    print("Congratulations! You've reached the target score and won the game!")

if __name__ == "__main__":
    play_game()
```
