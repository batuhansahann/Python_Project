```python
import time
import os
import math

def clear_screen():
    """Clear the console screen."""
    os.system('cls' if os.name == 'nt' else 'clear')

def draw_lighthouse(angle):
    """Draw a lighthouse and simulate the rotating light beam."""
    radius = 10
    light_char = "O"
    empty_char = " "
    
    for y in range(-radius, radius + 1):
        row = ""
        for x in range(-radius, radius + 1):
            distance = math.sqrt(x**2 + y**2)
            if radius - 0.5 < distance < radius + 0.5:  # Draw circle border
                row += "#"
            elif math.isclose(math.atan2(y, x), math.radians(angle), abs_tol=0.2):  # Light beam
                row += light_char
            else:
                row += empty_char
        print(row)

def lighthouse_simulation():
    """Simulate a rotating lighthouse."""
    print("Lighthouse Simulation: Press Ctrl+C to stop.")
    angle = 0

    try:
        while True:
            clear_screen()
            print(f"Lighthouse angle: {angle}°")
            draw_lighthouse(angle)
            angle = (angle + 15) % 360  # Increment angle, loop back at 360
            time.sleep(0.5)
    except KeyboardInterrupt:
        clear_screen()
        print("Lighthouse simulation stopped. Goodbye!")

if __name__ == "__main__":
    lighthouse_simulation()
```
