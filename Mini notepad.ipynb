```python
import os

def display_menu():
    print("\nMini Not Defteri")
    print("1 - Add a note")
    print("2 - View notes")
    print("3 - Save notes to file")
    print("4 - Exit")

def add_note():
    note = input("Enter your note: ")
    notes.append(note)
    print("Note added!")

def view_notes():
    if notes:
        print("\nYour Notes:")
        for idx, note in enumerate(notes, 1):
            print(f"{idx}. {note}")
    else:
        print("No notes yet. Add a note first!")

def save_notes():
    with open("notes.txt", "w") as file:
        for note in notes:
            file.write(note + "\n")
    print("Notes saved to notes.txt")

def load_notes():
    if os.path.exists("notes.txt"):
        with open("notes.txt", "r") as file:
            return [line.strip() for line in file]
    return []

def main():
    global notes
    notes = load_notes()  # Load existing notes from file if available

    while True:
        display_menu()
        choice = input("Choose an option: ")
        
        if choice == "1":
            add_note()
        elif choice == "2":
            view_notes()
        elif choice == "3":
            save_notes()
        elif choice == "4":
            print("Exiting the application. Goodbye!")
            break
        else:
            print("Invalid option. Please choose 1, 2, 3, or 4.")

if __name__ == "__main__":
    main()
```
