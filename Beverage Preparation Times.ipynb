```python
def display_menu():
    """Display the menu with available drinks."""
    print("Welcome to the Drink Preparation Time Information!")
    print("Please choose a drink to see the preparation time:\n")
    print("1 - Tea")
    print("2 - Coffee")
    print("3 - Espresso")
    print("4 - Latte")
    print("5 - Cappuccino")
    print("6 - Exit")

def get_preparation_time(drink_choice):
    """Return the preparation time for the chosen drink."""
    if drink_choice == 1:
        return "Tea: 5 minutes (boil water, steep tea bag, serve)"
    elif drink_choice == 2:
        return "Coffee: 4 minutes (brew coffee, serve)"
    elif drink_choice == 3:
        return "Espresso: 2 minutes (brew espresso)"
    elif drink_choice == 4:
        return "Latte: 5 minutes (brew espresso, steam milk, combine)"
    elif drink_choice == 5:
        return "Cappuccino: 6 minutes (brew espresso, steam milk, froth milk, combine)"
    else:
        return "Invalid choice. Please select a valid option."

def main():
    """Main function to run the program."""
    while True:
        display_menu()
        try:
            choice = int(input("Enter your choice (1-6): "))
            if choice == 6:
                print("Thank you! Have a great day!")
                break
            preparation_time = get_preparation_time(choice)
            print(f"\n{preparation_time}\n")
        except ValueError:
            print("Invalid input! Please enter a number between 1 and 6.\n")

if __name__ == "__main__":
    main()
```
